```python
#connecting web driver to the installed libaray
drive = webdriver.Chrome("C:/Users/Ankit anand/selenium/chromedriver.exe")
#commanding driver to open the url
drive.get('https://www.flipkart.com/')
time.sleep(5)
drive.maximize_window()
close=drive.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']")
close.click()
search=drive.find_element_by_xpath('//input[@class="_3704LK"]')
search.send_keys("pixel 4A")
cli=drive.find_element_by_xpath('//button[@class="L0Z3Pu"]')
cli.click()
```

```python
#creating empty list so it can pbe appended
Brand_Name=[]
Smartphone_name=[]
Colour=[]
RAM=[]
Storage=[]
Primary_Camera=[]
Secondary_Camera=[]
Display_Size=[]
Battery_Capacity=[]
Price=[]
Product_URL=[]
```

```python
urls=drive.find_elements_by_xpath('//a[@class="_1fQZEK"]')
for i in urls:
    Product_URL.append(i.get_attribute('href'))
for i in Product_URL[0:]:  
    drive.get(i)
    time.sleep(5)    
    try:
        Smart=drive.find_element_by_xpath("//span[@class='B_NuCI']")
        Smartphone_name.append(Smart.text)
    except NoSuchElementException:
        Smartphone_name.append("-")
    try:
        Pri=drive.find_element_by_xpath("//div[@class='_30jeq3 _16Jk6d']")
        Price.append(Pri.text)
    except NoSuchElementException:
        Price.append("-")
    try :
        Ra=drive.find_element_by_xpath("/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[1]")
        RAM.append(Ra.text.split('|')[0])
    except NoSuchElementException:
        Return_Exchange.append("-")
    try:
        rom=drive.find_element_by_xpath("/html/body/div[1]/div/div[3]/div[1]/div[2]/div[7]/div[1]/div/div[2]/ul/li[1]")
        Storage.append(rom.text.split('|')[1])
    except NoSuchElementException:
        Storage.append("-")
    try:
        Primary=drive.find_element_by_xpath("/html/body/div[1]/div/div[3]/div[1]/div[2]/div[7]/div[1]/div/div[2]/ul/li[3]")
        Primary_Camera.append(Primary.text.split('|')[0])
    except NoSuchElementException:
        Primary_Camera.append("-")
    try:
        Secondary=drive.find_element_by_xpath("/html/body/div[1]/div/div[3]/div[1]/div[2]/div[7]/div[1]/div/div[2]/ul/li[3]")
        Secondary_Camera.append(Secondary.text.split('|')[1])
    except:
        Secondary_Camera.append("-") 
    try:
        Battery=drive.find_element_by_xpath("/html/body/div[1]/div/div[3]/div[1]/div[2]/div[7]/div[1]/div/div[2]/ul/li[4]")
        Battery_Capacity.append(Battery.text)
    except :
        Battery_Capacity.append("-")         
    try:
        Display=drive.find_element_by_xpath("/html/body/div[1]/div/div[3]/div[1]/div[2]/div[7]/div[1]/div/div[2]/ul/li[2]")
        Display_Size.append(Display.text)
    except :
        Display_Size.append("-")
    try:
        Colo=drive.find_element_by_xpath("/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div/div[1]/table/tbody/tr[4]/td[2]/ul/li")
        Colour.append(Colo.text)
    except:
        Colour.append("-")            
```
